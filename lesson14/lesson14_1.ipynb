```python
import requests


#需要安裝外部套件 pip install requests
#requests可以使用下列HTTP六個方法，使用後會傳回一個Response物件
#六個方法裡，GET和POST是最常用的兩個方法，
#GET提交的參數會放在標頭中傳送(public)
#POST則是放在內容中傳送(private)
#GET: requests.get(url)向指定資源提交請求，可額外設定 params 參數字典。
#POSTT: requests.post(url)向指定資源提交請求，可額外設定 data 參數字典。
#PUTT: requests.put(url)向指定資源提供最新內容，可額外設定 data 參數字典。
#DELETET: requests.delete(url)請求刪除指定的資源。
#HEADT: requests.head(url)請求提供資源的回應標頭(不含內容)
#OPTIONST: requests.options(url)請求伺服器提供資源可用的功能選項



#一個csv的下載網址
url:str = "https://www.moeaea.gov.tw/ECW/populace/opendata/wHandOpenData_File.ashx?set_id=248"

#text1(url,params=None,**kwargs沒有限定數量的引數呼叫)
response:requests.Response = requests.get(url)
```

```python
#檢查是否下載成功(status_code == 200)
if response.status_code==200:
    print("下載成功")
    s:str = response.text
    display(s)

    #擷取第一行
    row_data=s.split("\r\n")
    display(row_data)

    #寫入csv檔
    fileName="download"
    with open(fileName+".csv",mode="w",encoding="utf-8") as f1:
        f1.write(s)
    print("寫入csv檔成功")

elif response.status_code>=400:
    print("下載失敗")
elif response.status_code>200:
    print("下載有問題")
```
